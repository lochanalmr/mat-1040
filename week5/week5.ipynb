```python
def highest_score(scores_list):
    comparison_score = scores_list[0]
    for score in scores_list:
        if comparison_score < score:
            comparison_score = score
    max_score = comparison_score
    return max_score

scores_list = [80,60,50,65,75,55]
print(highest_score(scores_list))
```

```python
def minimum_score(scores_list):
    comparison_score = scores_list[0]
    for score in scores_list:
        if comparison_score > score:
            comparison_score = score
    min_score = comparison_score
    return min_score

scores_list = [80,60,50,65,75,55]
print(minimum_score(scores_list))
```

```python
def integer_values(values_list):
    final_string = ""
    for value in values_list:
        if isinstance(value, int):
            final_string += str(value)
    return final_string

print(integer_values(values_list = [11,30.1,90.2,30,45.1,54,'54']))
```

```python
def sum_score_above_average(student_scores):
    sum = 0
    count = 0
    for score in student_scores:
        sum += score
        count += 1
    average = sum / count

    sum_above_average = 0
    for score in student_scores:
        if score > average:
            sum_above_average += score
    return sum_above_average

print(sum_score_above_average(student_scores = [80,60,50,65,75,55]))
```

```python
def count_passed(marks):
    count = 0
    for mark in marks:
        if mark >= 50:
            count += 1
    return count
print(count_passed(marks = [45,50,80,30,60]))
```

```python
def shopping_total(prices):
    total_price = 0
    for price in prices:
        total_price += price
    return total_price
print(shopping_total(prices = [100,250,50]))
```

```python
def strong_password(password):
    count = 0
    digit_check = False
    for character in password:
        count += 1
        if character.isdigit():
            digit_check = True
    if count >= 8 and digit_check == True:
        return True
    else:
        return False

password = 'hello123'
strong_password(password)
```

```python
def true_love_score(name1, name2):
    name1 = name1.lower()
    name2 = name2.lower()
    name = name1 + name2
    letter_set_true = ['t', 'r', 'u', 'e']
    count_true = 0
    for letter in name:
        if letter in letter_set_true:
            count_true += 1

    letter_set_love = ['l', 'o', 'v', 'e']
    count_love = 0
    for letter in name:
        if letter in letter_set_love:
            count_love += 1
    return int(str(count_true) + str(count_love))

name1 = "Kanye West"
name2 = "Kim Kardashian"
print(true_love_score(name1, name2))
```

```python
def multiplication_table(x):
    multiplication_table = []
    for i in range(10):
        y = x * (i + 1)
        multiplication_table.append(y)
    return multiplication_table
print(multiplication_table(5))
```

```python
def atm_withdraw(balance, amount):
    if amount > 0 and amount <= balance:
        return "Approved"
    else:
        return "Declined"
print(atm_withdraw(balance = 5000, amount = 1200))
```
