```python
def simple_calculator(num1, num2, operator):
    if operator == "+":
        return num1 + num2
    elif operator == "-":
        return num1 - num2
    elif operator == "*":
        return num1 * num2
    elif operator == "/":
        return num1 / num2
    return None

num1 = float(input())
num2 = float(input())
operator = input()
print(simple_calculator(num1, num2, operator))
```

Question 14

```python
def triangle_type(a, b, c):
    if not (a + b > c) and not (a + c > b) and not (b + c > a):
        return "Invalid triangle"
    elif a == b and b == c:
        return "Equilateral"
    elif a == b or b == c or a == c:
        return "Isosceles"
    else:
        return "Scalene"

a = int(input())
b= int(input())
c = int(input())
print(triangle_type(a, b, c))
```

Question 15

```python
def discriminant_checker(a, b, c):
    D = b ** 2 - 4 * a * c
    if D > 0:
        return "Two real roots"
    elif D == 0:
        return "One real root"
    elif D < 0:
        return "No real roots"

a = int(input())
b = int(input())
c = int(input())
print(discriminant_checker(a, b, c))
```

Question 16

```python
def electricity_bill(units):
    if units <= 100:
        return units * 20
    elif units <= 200:
        return 100 * 20 + (units - 100) * 30
    else:
        return 100 * 20 + 100 * 30 + (units - 200) * 45

units = int(input())
print(electricity_bill(units))
```

Question 17

```python
def taxi_fare(km):
    if km <= 1:
        return 150
    else:
        return 150 + (km - 1) * 80

km = float(input())
print(taxi_fare(km))
```

Question 18

```python
def student_result(m1, m2, m3):
    if m1 >= 35 and m2 >= 35 and m3 >= 35:
        avg_mark = (m1 + m2 + m3) / 3
        if avg_mark >= 50:
            return "Pass"
        else:
            return "Fail"
    else:
        return "Fail"

m1 = float(input())
m2 = float(input())
m3 = float(input())
print(student_result(m1, m2, m3))
```
